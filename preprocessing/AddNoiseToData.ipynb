```python
import matplotlib.pyplot as plt
from skimage.util import random_noise
from skimage import data, img_as_float
import matplotlib.image as mpimg
import os
```

```python
def NoisifyImage(imagePath, sigma):
    image = mpimg.imread(imagePath,format='jpg')
    image = img_as_float(image)
    return random_noise(image, var=sigma**2)
```

```python
def SaveImage(image, outputPath):
    mpimg.imsave(outputPath, image)
```

```python
def AddNoiseToImages(input_directory,output_directory,sigma):
    for filename in os.listdir(input_directory):
        inputFilePath = os.path.join(input_directory, filename)
        noisiyImage = NoisifyImage(inputFilePath,sigma)
        
        outputFilePath = os.path.join(output_directory, filename + '_noise' + str(sigma)[2:]+'.png') #ignores the dot for the sigma
        mpimg.imsave(outputFilePath, noisiyImage)
```

```python
pathTrainHires = '/home/Matthew/image-super-resolution/data/train/hires/'
pathTrainLowres = '/home/Matthew/image-super-resolution/data/train/lowres/'
pathValidHires = '/home/Matthew/image-super-resolution/data/valid/hires/'
pathValidLowres = '/home/Matthew/image-super-resolution/data/valid/lowres/'
```

```python
AddNoiseToImages(pathTrainHires,pathTrainLowres,0.115)
```

```python
AddNoiseToImages(pathValidHires,pathValidLowres,0.115)
```
